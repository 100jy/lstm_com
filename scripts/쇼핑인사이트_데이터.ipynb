```python
import numpy as np
import pandas as pd
form = "C:/Users/wnduq/Desktop/쇼핑인사이트/성별_연령별_쇼핑인사이트/{}대_{}성_{}.csv" 

a=np.array([1,2,3,4,5,6])
a=a*10

b=["남","여"]

c=["출산&식품","패션잡화&화장품","디지털&가구","스포츠&생활","여행&면세점"]

data=pd.DataFrame(np.zeros((365,1)))

for i in a :
    for j in b:
        for k in c:
            x=pd.read_csv(form.format(i,j,k),encoding="euc-kr")["adj_ratio"]
            y=pd.DataFrame(np.array(x[365:365+365]))
            z=pd.DataFrame(np.array(x[365+365:]))
            data=pd.concat([data,y,z],axis=1)
            

data

```
